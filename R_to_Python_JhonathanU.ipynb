## Part A -- Translation (Conceptual)

For the R code below:

2.  Translate it into Python

``` r
x <- c(1, 3, 5, 7)

mean(y)
```

Clearly show your Python code and final result.

------------------------------------------------------------------------

````python
## I'm thinking the code will see if the elements of the vector stored in x is actually greater than 4. If it is, than what it will do is square it. On the other hand if it is not larger than 4,\n than it will add 1 to that element instead. After that it would take the mean of the new array which would now be stored in y variable.
## Knowing this I think what will be printed out is [2, 4, 25, 49] and the mean for these numbers is 20.

## Python Code
import numpy as np

x = np.array([1, 3, 5, 7])
y = np.where(x > 4, x ** 2, x + 1)

print(y)
````
